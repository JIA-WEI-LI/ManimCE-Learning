```python
#@title Logo
#@markdown #  <b>ManimCE Logo Animation</b>
#@markdown This is a ManimCE logo animation designed by myself ( only have def ),&ensp;
#@markdown you can choose whether to join or not in the bottom part.
from manim import *
def ManimCELogo(self):
  # logo
  banner = ManimBanner(dark_theme=False)
  circle_banner = Circle(color="#ece6e2", fill_opacity=1).surround(banner).scale(2)
  # logo Text
  logo_Text = Text("Manim Community", font_size=48).shift(RIGHT)
  logo_MadeBy = Text("Made by", font_size=24).next_to(logo_Text, UP).shift(LEFT*2.5)
  logo_Web = Text("https://docs.manim.community/en/stable/index.html", font_size=16).next_to(logo_Text, DOWN)
  # Group
  banner_Group = VGroup(circle_banner, banner)
  logo_Text_Group = Group(logo_MadeBy, logo_Web)

  self.play(FadeIn(circle_banner), banner.create())
  self.play(banner.animate.scale(0.2), circle_banner.animate.scale(0.1))
  self.play(banner_Group.animate.next_to(logo_Text_Group, LEFT))
  self.play(FadeIn(logo_Text_Group), Write(logo_Text), run_time=2.5)
  self.wait(3)
  self.play(FadeOut(banner_Group), FadeOut(logo_Text_Group), FadeOut(logo_Text))
  self.wait()
```

```python
#@title Code Setting
def ChangeToManimDirection(RL=str, RL_u=float, UD=str, UD_u=float):
  d = [0, 0, 0]
  if RL == "LEFT":
    d[0] = -1*RL_u    
  elif RL == "RIGHT":
    d[0] = 1*RL_u
  else:
    print(f"ERROR: Direction _LR string is not \"LEFT\" or \"RIGHT\", check Setting again.")
  if UD == "UP":
    d[1] = 1*UD_u 
  elif UD == "DOWN":
    d[1] = -1*UD_u
  else:
    print(f"ERROR: Direction _UD string is not \"LEFT\" or \"RIGHT\", check Setting again.")
  return d
```

## Create and Move a Rectangle

```python
#@markdown # **Custom Animation Content**
#@markdown You can use this template as the main sample to customize your own animation content, try to adjust the color (according to the hexadecimal color code table) or size,
#@markdown and even adjust the desired moving position of each step.

#@markdown If you have any questions, you can refer to my Github
#@markdown [Create And Move A Rectangle](https://github.com/JIA-WEI-LI/ManimCE-Learning/blob/d279c68c08cb69c7bb9592f4878e76f8fa2381e4/%E4%B8%AD%E6%96%87%E7%89%88/Tutorial%201/Create%20And%20Move%20A%20Rectangle.md)
Run_time = list(range(4))
#@markdown ---
#@markdown ## <font color="#dfdf88">**Rectangle Setting**</font>
#@markdown The following are the basic settings about the rectangle, which are usually adjusted inside the `Rectangle()` function.
#@markdown ###  <font color="#efbcbc">**Color**</font>
#@markdown Here you can adjust the line color about the rectangle ( `color` ), opacity of stroke line color (`stroke_opacity`) , the fill color inside the rectangle ( `fill_color` ), the fill color transparency inside the rectangle ( `fill_opacity` )
#@markdown > Color reference[Manim Community v0.16.0.post0 -- Colors](https://docs.manim.community/en/stable/reference/manim.utils.color.Colors.html#manim.utils.color.Colors)
# Initial setting#83C167#E65A4C0.711
Stroke_Color = "#83C167" #@param {type:"string"}
Stroke_Opacity = 0.7 #@param {type:"slider", max:1, min:0, step:0.1}
Fill_Color = "#E65A4C" #@param {type:"string"}
Fill_Opacity = 0.5 #@param {type:"slider", max:1, min:0, step:0.1}
#@markdown ### <font color="#efbcbc"> **Height & Width**</font>
#@markdown Here you can adjust the `length` and `width` of the rectangle (the default is 1)
Height = 1 #@param {type:"slider", max:2, min:0, step:0.1}
Width = 1 #@param {type:"slider", max:2, min:0, step:0.1}
#@markdown ---
#@markdown ## <font color="#dfdf88">**Rectangle Animation Setting**</font>
#@markdown The following are the details of adjusting the movement animation of the rectangle.
#@markdown There are four steps in total that can be modified. Each step is divided into two options: `UP` `DOWN` and `LEFT` `RIGHT` , 
#@markdown  and each of them are multiplied by their own **Moving unit**.

#@markdown For Example , *Move 2 units to the right and 3 units down* equal to
#@markdown `RIGHT * 2 + DOWN * 3`
#@markdown ## 1 <font color="#69afff">**First Step**</font>
First_Step_Direction_RL = "RIGHT" #@param ['LEFT', 'RIGHT']
First_Step_Unit_RL = 2 #@param {type:"slider", max:5, min:0, step:0.1}
First_Step_Direction_UD = "UP" #@param ['UP', 'DOWN']
First_Step_Unit_UD = 0 #@param {type:"slider", max:5, min:0, step:0.1}
Run_time[0] = 2 #@param {type:"slider", max:5, min:1, step:0.1}
#@markdown ## 2 <font color="#69afff">**Second Step**</font>
Second_Step_Direction_RL = "RIGHT" #@param ['LEFT', 'RIGHT']
Second_Step_Unit_RL = 0 #@param {type:"slider", max:5, min:0, step:0.1}
Second_Step_Direction_UD = "UP" #@param ['UP', 'DOWN']
Second_Step_Unit_UD = 3 #@param {type:"slider", max:5, min:0, step:0.1}
Run_time[1] = 2 #@param {type:"slider", max:5, min:1, step:0.1}
#@markdown ## 3 <font color="#69afff">**Third Step**</font>
Third_Step_Direction_RL = "LEFT" #@param ['LEFT', 'RIGHT']
Third_Step_Unit_RL = 5 #@param {type:"slider", max:5, min:0, step:0.1}
Third_Step_Direction_UD = "DOWN" #@param ['UP', 'DOWN']
Third_Step_Unit_UD = 5 #@param {type:"slider", max:5, min:0, step:0.1}
Run_time[2] = 2 #@param {type:"slider", max:5, min:1, step:0.1}
#@markdown ## 4 <font color="#69afff">**Forth Step**</font>
Forth_Step_Direction_RL = "RIGHT" #@param ['LEFT', 'RIGHT']
Forth_Step_Unit_RL = 1 #@param {type:"slider", max:5, min:0, step:0.1}
Forth_Step_Direction_UD = "UP" #@param ['UP', 'DOWN']
Forth_Step_Unit_UD = 1.5 #@param {type:"slider", max:5, min:0, step:0.1}
Run_time[3] = 2 #@param {type:"slider", max:5, min:1, step:0.1}

d=list(range(4))
d[0]=ChangeToManimDirection(First_Step_Direction_RL, First_Step_Unit_RL, First_Step_Direction_UD, First_Step_Unit_UD)
d[1]=ChangeToManimDirection(Second_Step_Direction_RL, Second_Step_Unit_RL, Second_Step_Direction_UD, Second_Step_Unit_UD)
d[2]=ChangeToManimDirection(Third_Step_Direction_RL, Third_Step_Unit_RL, Third_Step_Direction_UD, Third_Step_Unit_UD)
d[3]=ChangeToManimDirection(Forth_Step_Direction_RL, Forth_Step_Unit_RL, Forth_Step_Direction_UD, Forth_Step_Unit_UD)
```

```python
%%manim -qm -v WARNING CreateAndMoveARectangle
class CreateAndMoveARectangle(Scene):
  def construct(self):
    # Add NumberPlane with Number label
    ax = Axes(x_range=[-8, 8, 1], y_range=[-6, 6, 1], x_length=16, y_length=12).add_coordinates()
    number_plane = NumberPlane(background_line_style={
        "stroke_color": BLUE,
        "stroke_width": 4,
        "stroke_opacity": 0.6,
    })
    self.add(ax, number_plane)

    # Add Rectangle
    box = Rectangle(color=Stroke_Color, stroke_opacity=Stroke_Opacity, fill_color=Fill_Color, fill_opacity=Fill_Opacity, height=Height, width=Width)

    self.add(box)
    for i in range(4):
      self.play(box.animate.shift(d[i]), run_time=Run_time[i])
```

## Create Axis and Transform Object

```python
#@markdown Because colab cannot input three-dimensional coordinates in one line,
#@markdown you need to enter the coordinate values yourself and separate them with commas.
X_Range = [ -3, 3, 1] #@param
Y_Range = [ -3, 3, 1] #@param
X_Length = 6 #@param {type:"number"}
Y_Length = 6 #@param {type:"number"}
#@markdown ## <font color="#ffffff">**Axis Setting**</font>
Axes_Position_Direction_LR = "LEFT" #@param ['LEFT','RIGHT']
Axes_Position_Unit_LR = 0.5 #@param {type:"slider", min:0, max:10, step:0.1}
Axes_Position_Direction_UD = "UP" #@param ['UP','DOWN']
Axes_Position_Unit_UD = 0 #@param {type:"slider", min:0, max:10, step:0.1}
#@markdown ## <font color="#ffffff">**Circle Setting**</font>
#@markdown ### **Circle Stroke & Color**
#@markdown > Color reference[Manim Community v0.16.0.post0 -- Colors](https://docs.manim.community/en/stable/reference/manim.utils.color.Colors.html#manim.utils.color.Colors)
Circle_Stroke_Width = 6 #@param {type:"slider", min:0, max:10, step:0.1}
Circle_Stroke_Color = "#FFFF00" #@param {type:"string"}
Circle_Stroke_Opacity = 1 #@param {type:"slider", min:0, max:1, step:0.1}
Circle_Fill_Color = "#FC6255" #@param {type:"string"}
Circle_Fill_Opacity = 0.8 #@param {type:"slider", min:0, max:1, step:0.1}
#@markdown ### **Circle Position**
Circle_Position_Direction_LR = "RIGHT" #@param ['LEFT','RIGHT']
Circle_Position_Unit_LR = 3 #@param {type:"slider", min:0, max:10, step:0.1}
Circle_Position_Direction_UD = "DOWN" #@param ['UP','DOWN']
Circle_Position_Unit_UD = 3 #@param {type:"slider", min:0, max:10, step:0.1}

#@markdown ## <font color="#ffffff">**Triangle Setting**</font>
#@markdown ### **Triangle Stroke & Color**
Triangle_Stroke_Color = "#FF862F" #@param {type:"string"}
Triangle_Stroke_Opacity = 0.8 #@param {type:"slider", min:0, max:1, step:0.1}
Triangle_Fill_Color = "#888888" #@param {type:"string"}
Triangle_Fill_Opacity = 1 #@param {type:"slider", min:0, max:1, step:0.1}
#@markdown ### **Triangle Position**
Triangle_Position_Direction_LR = "RIGHT" #@param ['LEFT','RIGHT']
Triangle_Position_Unit_LR = 3 #@param {type:"slider", min:0, max:10, step:0.1}
Triangle_Position_Direction_UD = "DOWN" #@param ['UP','DOWN']
Triangle_Position_Unit_UD = 3 #@param {type:"slider", min:0, max:10, step:0.1}

axes_pos = ChangeToManimDirection(Axes_Position_Direction_LR, Axes_Position_Unit_LR, Axes_Position_Direction_UD, Axes_Position_Unit_UD)
tri_pos = ChangeToManimDirection(Triangle_Position_Direction_LR, Triangle_Position_Unit_LR, Triangle_Position_Direction_UD, Triangle_Position_Unit_UD)
```

```python
%%manim -qm -v WARNING CreateAxisAndTransformObject
class CreateAxisAndTransformObject(Scene):
  def construct(self):
    axes = Axes(
        x_range=X_Range, y_range=Y_Range,
        x_length=X_Length, y_length=Y_Length)
    axes.to_edge(axes_pos)

    circle = Circle(stroke_width=Circle_Stroke_Width, color=Circle_Stroke_Color, stroke_opacity=Circle_Stroke_Opacity, fill_color=Circle_Fill_Color, fill_opacity=Circle_Fill_Opacity)
    circle.set_width(2).to_edge(DR, buff=2)
    triangle = Triangle(color=Triangle_Stroke_Color, stroke_opacity=Triangle_Stroke_Opacity, fill_color=Triangle_Fill_Color, fill_opacity=Triangle_Fill_Opacity).set_height(2).shift(tri_pos)

    self.play(Write(axes))
    self.play(DrawBorderThenFill(circle))
    self.play(circle.animate.set_width(1))
    self.play(Transform(circle, triangle), run_time=3)
```

## MathTex and Updater 

```python
#@markdown ## <font color="#dfdf88">**Rectangle Setting**</font>
#@markdown ###  <font color="#efbcbc">**Rectangle Stroke and Color**</font>
Rectangle_Stroke_Width = 8 #@param {type:"number"}
Rectangle_Stroke_Color = "#FFFFFF" #@param {type:"string"}
Rectangle_Fill_Color = "#9CDCEB" #@param {type:"string"}
Rectangle_Width = 4.5 #@param {type:"slider", min:0, max:10, step:0.1}
Rectangle_Height = 2 #@param {type:"slider", min:0, max:10, step:0.1}
#@markdown ### **Rectangle Position**
Rectangle_Position_Direction_LR = "LEFT" #@param ['LEFT','RIGHT']
Rectangle_Position_Unit_LR = 1 #@param {type:"slider", min:0, max:5, step:0.1}
Rectangle_Position_Direction_UD = "UP" #@param ['UP','DOWN']
Rectangle_Position_Unit_UD = 3 #@param {type:"slider", min:0, max:5, step:0.1}

Rectangle_Pos = ChangeToManimDirection(Rectangle_Position_Direction_LR, Rectangle_Position_Unit_LR, Rectangle_Position_Direction_UD, Rectangle_Position_Unit_UD)

#@markdown ## <font color="#dfdf88">**MathTex**</font>
MathText_by_LATEX = "\\frac{3}{4}=0.75" #@param {type:"string"}
MathText_Color_begin = "#83C167" #@param {type:"string"}
MathText_Color_end = "#D147BD" #@param {type:"string"}
MathText_Scale_Height = 1.5 #@param {type:"slider", min:0, max:10, step:0.1}

#@markdown ---
#@markdown ## <font color="#dfdf88">**Animation Setting**</font>
Animate_Runtime = 6
#@markdown ### **Rectangle with MathText Move**
All_Animation_Direction_LR = "LEFT" #@param ['LEFT','RIGHT']
All_Animation_Unit_LR = 1 #@param {type:"slider", min:0, max:5, step:0.1}
All_Animation_Direction_UD = "UP" #@param ['UP','DOWN']
All_Animation_Unit_UD = 1 #@param {type:"slider", min:0, max:5, step:0.1}
#@markdown ### **Rectangle Move**
Rectangle_Animation_Direction_LR = "LEFT" #@param ['LEFT','RIGHT']
Rectangle_Animation_Unit_LR = 1 #@param {type:"slider", min:0, max:5, step:0.1}
Rectangle_Animation_Direction_UD = "UP" #@param ['UP','DOWN']
Rectangle_Animation_Unit_UD = 1 #@param {type:"slider", min:0, max:5, step:0.1}

All_Animation_Pos = ChangeToManimDirection(All_Animation_Direction_LR, All_Animation_Unit_LR, All_Animation_Direction_UD, All_Animation_Unit_UD)
Rectangle_Animation_Pos = ChangeToManimDirection(Rectangle_Animation_Direction_LR, Rectangle_Animation_Unit_LR, Rectangle_Animation_Direction_UD, Rectangle_Animation_Unit_UD)
```

```python
%%manim -qm -v WARNING MathTexAndUpdater
class MathTexAndUpdater(Scene):
  def construct(self):
    rectangle = RoundedRectangle(
        stroke_width=Rectangle_Stroke_Width, 
        color=Rectangle_Stroke_Color, 
        fill_color=Rectangle_Fill_Color, 
        width=Rectangle_Width, 
        height=Rectangle_Height ).shift([2, 0, 0])

    mathText = MathTex(MathText_by_LATEX).set_color_by_gradient(MathText_Color_begin, MathText_Color_end).set_height(MathText_Scale_Height)
    mathText.move_to(rectangle.get_center())
    mathText.add_updater(lambda x:x.move_to(rectangle.get_center()))

    self.play(FadeIn(rectangle))
    self.play(Write(mathText))
    self.play(rectangle.animate.shift(All_Animation_Pos), run_time=Animate_Runtime)
    self.wait()
    mathText.clear_updaters()
    self.play(rectangle.animate.shift(Rectangle_Animation_Pos), run_time=Animate_Runtime)
```
